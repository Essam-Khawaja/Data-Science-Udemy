```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
rawData = pd.read_csv('1.03. Dummies.csv')
```

```python
rawData
```

```python
data = rawData.copy()
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No':0})
data
```

```python
data.describe()
```

## Regression

```python
y = data["GPA"]
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
model = sm.OLS(y, x).fit()
model.summary()
```

```python
# Now, we have two cases for attendance. One is when they attend, meaning that we apply its coefficient, and another where they do not attend so we do not apply the coefficient
# Here are the two equations:
yHat_no = 0.6439 + 0.0014*data['SAT']   # GPA = 0.6439 + 0.0014*SAT + 0.2226*0
yHat_yes = (0.6439 + 0.2226) + 0.0014*data['SAT']   # GPA = 0.6439 + 0.0014*SAT + 0.2226*1

plt.scatter(data['SAT'], y)
fig = plt.plot(data['SAT'], yHat_no, c='orange')
fig = plt.plot(data['SAT'], yHat_yes, c='blue')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

## How to make predictions based on regression

```python
x
```

```python
# We want to predict GPA for the following cases: 1700 with no attendance, 1670 with attendance.
# Hence, we create a new data frame with those cases within it. Think of this as our test data, and our earlier data as training.

new_data = pd.DataFrame({
    'const': 1,
    'SAT':[1700, 1670],
    'Attendance': [0,1]
})

new_data = new_data[['const', 'SAT', "Attendance"]]
new_data
```

```python
predictions = model.predict(new_data)
predictions
```
